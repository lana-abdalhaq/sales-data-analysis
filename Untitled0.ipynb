```python
import pandas as pd

df = pd.read_csv("Amazon Sale Report.csv")

df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df = df.drop(columns=["index", "Unnamed: 22"])

df.shape
```

```python
df["Date"] = pd.to_datetime(df["Date"])

df["Date"].dtype
```

```python
df[["Amount", "currency", "Courier Status"]].isnull().sum()
```

```python
df[df["Amount"].isnull()]["Status"].value_counts()
```

```python
df = df[df["Status"] != "Cancelled"]

df.shape
```

```python
df["Amount"].isnull().sum()
```

```python
df[df["Amount"].isnull()]["Status"].value_counts()
```

```python
df = df[df["Amount"].notnull()]

df.shape
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()

df.shape
```

```python
(df["Amount"] < 0).sum()
```

```python
df["Amount"].sum()
```

```python
df["Qty"].sum()
```

```python
df["Amount"].sum() / df["Qty"].sum()
```

```python
df.groupby("Category")["Amount"].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

category_sales = df.groupby("Category")["Amount"].sum().sort_values(ascending=False)

category_sales.plot(kind="bar")
plt.title("Total Sales by Category")
plt.ylabel("Total Revenue")
plt.show()
```

```python
(category_sales / category_sales.sum()) * 100
```

```python
df["Month"] = df["Date"].dt.to_period("M")
```

```python
df[["Date", "Month"]].head()
```

```python
monthly_sales = df.groupby("Month")["Amount"].sum()

monthly_sales
```

```python
import matplotlib.pyplot as plt

plt.figure()
monthly_sales.plot(kind="line")
plt.title("Monthly Sales Trend")
plt.ylabel("Total Revenue")
plt.show()
```

```python
state_sales = df.groupby("ship-state")["Amount"].sum().sort_values(ascending=False)

state_sales.head(10)
```

**Executive Summary**

This project analyzed Amazon e-commerce sales data to understand performance across products, time, and geography.
After cleaning the dataset (removing cancelled orders, missing revenue records, and duplicates), key business KPIs were calculated:

Total Revenue: 71,670,759

Total Orders: 103,070

Total Units Sold: 110,823

Average Order Value (AOV): ~695.36

Average Revenue per Unit: ~646.71

Key findings:

Revenue is highly concentrated in a few categories: Set (~49.85%) and Kurta (~27.10%), together generating ~77% of total revenue.

Monthly sales peaked in April 2022, followed by a gradual decline in May and June.

The top revenue states are Maharashtra and Karnataka, indicating geographical concentration in major markets.

Order amounts are right-skewed, with most orders in the mid-range and a small number of high-value outliers.

The return rate is low (~1.77%), suggesting good product quality and customer satisfaction.

Overall, the business shows strong performance driven by a limited set of product categories and key regions, with opportunities to improve sales stability and increase average basket size through bundling and targeted marketing.
